```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
train_path = "Problem 2/train_happy.h5"
test_path = "Problem 2/test_happy.h5"

with h5py.File(train_path, "r") as train_dataset:
    X_train = np.array(train_dataset["train_set_x"][:])
    Y_train = np.array(train_dataset["train_set_y"][:])

with h5py.File(test_path, "r") as test_dataset:
    X_test = np.array(test_dataset["test_set_x"][:])
    Y_test = np.array(test_dataset["test_set_y"][:])

print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)
print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)
```

```python
Y_train = Y_train.reshape(-1)
Y_test = Y_test.reshape(-1)

print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
plt.figure(figsize=(12,6))

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(X_train[i])
    plt.title(f"Label: {Y_train[i]}")
    plt.axis("off")

plt.show()
```

```python
num_smiling_train = np.sum(Y_train == 1)
num_non_smiling_train = np.sum(Y_train == 0)

num_smiling_test = np.sum(Y_test == 1)
num_non_smiling_test = np.sum(Y_test == 0)

print("Training set:")
print("Smiling:", num_smiling_train)
print("Not smiling:", num_non_smiling_train)

print("\nTest set:")
print("Smiling:", num_smiling_test)
print("Not smiling:", num_non_smiling_test)
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0

X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

print("Flattened training shape:", X_train_flat.shape)
print("Flattened test shape:", X_test_flat.shape)
```

```python
#PART 2
```

```python
model_1 = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_flat.shape[1],)),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_1.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_1.summary()
```

```python
history_1 = model_1.fit(
    X_train_flat, Y_train,
    epochs=30,
    batch_size=32,
    validation_data=(X_test_flat, Y_test),
    verbose=1
)
```

```python
test_loss_1, test_acc_1 = model_1.evaluate(X_test_flat, Y_test, verbose=0)

print("Baseline ANN Test Loss:", test_loss_1)
print("Baseline ANN Test Accuracy:", test_acc_1)
```

```python
plt.figure(figsize=(8,5))
plt.plot(history_1.history['accuracy'], label='Train Accuracy')
plt.plot(history_1.history['val_accuracy'], label='Test Accuracy')
plt.title('Baseline ANN Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(history_1.history['loss'], label='Train Loss')
plt.plot(history_1.history['val_loss'], label='Test Loss')
plt.title('Baseline ANN Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#PART 3 added dropout
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model_2 = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_flat.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model_2.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_2.summary()
```

```python
history_2 = model_2.fit(
    X_train_flat, Y_train,
    epochs=30,
    batch_size=32,
    validation_data=(X_test_flat, Y_test),
    verbose=1
)
```

```python
test_loss_2, test_acc_2 = model_2.evaluate(X_test_flat, Y_test, verbose=0)

print("Improved ANN Test Loss:", test_loss_2)
print("Improved ANN Test Accuracy:", test_acc_2)
```

```python
plt.figure(figsize=(8,5))

plt.plot(history_1.history['val_accuracy'], label='Baseline Test Accuracy')
plt.plot(history_2.history['val_accuracy'], label='Improved Model Test Accuracy')

plt.title('Model Comparison - Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(history_2.history['loss'], label='Train Loss (Dropout Model)')
plt.plot(history_2.history['val_loss'], label='Test Loss (Dropout Model)')

plt.title('Dropout Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(history_1.history['val_loss'], label='Baseline Test Loss')
plt.plot(history_2.history['val_loss'], label='Dropout Test Loss')

plt.title('Loss Comparison: Baseline vs Dropout')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.show()
```

```python
model_3 = Sequential([
    Dense(256, activation='relu', input_shape=(X_train_flat.shape[1],)),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_3.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_3.summary()
```

```python
history_3 = model_3.fit(
    X_train_flat, Y_train,
    epochs=30,
    batch_size=32,
    validation_data=(X_test_flat, Y_test),
    verbose=1
)
```

```python
test_loss_3, test_acc_3 = model_3.evaluate(X_test_flat, Y_test, verbose=0)

print("Model 3 Test Loss:", test_loss_3)
print("Model 3 Test Accuracy:", test_acc_3)
```

```python
model_4 = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_flat.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_4.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_4.summary()
```

```python
history_4 = model_4.fit(
    X_train_flat, Y_train,
    epochs=30,
    batch_size=32,
    validation_data=(X_test_flat, Y_test),
    verbose=1
)
```

```python
test_loss_4, test_acc_4 = model_4.evaluate(X_test_flat, Y_test, verbose=0)

print("Model 4 Test Loss:", test_loss_4)
print("Model 4 Test Accuracy:", test_acc_4)
```
